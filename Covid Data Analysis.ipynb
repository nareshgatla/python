```python
import pandas as pd
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df = df.iloc[:-1, 1:]
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
total_states = np.arange(len(df['state_name']))
total_states
```

```python
 df['active']
```

```python
max(df['active'])
```

```python
max(df['positive'])
```

###Total positive cases based on states in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w',edgecolor='k')
plt.barh(total_states,df['positive'],align='center',alpha=0.5, color=(1,0,0),edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states,df['state_name'])
plt.xlim(1, max(df['positive'])+100)
plt.xlabel('Positive number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

###Total number of active cases based on states in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w',edgecolor='k')
plt.barh(total_states,df['active'],align='center',alpha=0.5, color=(1,0.4,0),edgecolor=(0.5,0.4,0.8))

plt.yticks(total_states,df['state_name'])
plt.xlim(1, max(df['active'])+100)
plt.xlabel('Positive number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

###Total number of deaths based on states in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w',edgecolor='k')
plt.barh(total_states,df['death'],align='center',alpha=0.5, color=(1,0.7,0),edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states,df['state_name'])
plt.xlim(1, max(df['death'])+100)
plt.xlabel('Positive number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

###Total new active cases based on states in india

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w',edgecolor='k')
plt.barh(total_states,df['new_active'],align='center',alpha=0.5, color=(0.4,0.6,0),edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states,df['state_name'])
plt.xlim(1, max(df['new_active'])+100)
plt.xlabel('Positive number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```
